## Question 3

## Solution:

```python
import numpy as np

# Compute ASCII sum of initials
x = ord("S")
y = ord("K")
xy = x + y  # Sum of ASCII values

print("X (ASCII sum):", xy)

# Part a: Create sales dataset
sales = np.array([xy, xy + 50, xy + 100, xy + 150, xy + 200])
print("Sales:", sales)

```

```python
# Part b: Compute tax rate and apply it
tax_rate = ((xy % 5) + 5) / 100
sales_with_tax = sales * (1 + tax_rate)
print("Sales with tax:", sales_with_tax)
```

```python
# Part c: Apply discounts
discounted_sales = np.where(sales < xy + 100, sales * 0.95, sales * 0.90)
print("Discounted sales:", discounted_sales)
```

```python
# Part d: Expand sales for multiple weeks
weekly_sales = np.tile(sales, (3, 1))  # Create a 3x5 matrix
weekly_sales = weekly_sales * (1 + 0.02) ** np.arange(3).reshape(-1, 1)  # Increase by 2% per week
print("Weekly sales data:\n", weekly_sales)
```

## Question 4

## Solution:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from -10 to 10 with 100 points
x = np.linspace(-10, 10, 100)

# Compute y values for each function
y1 = x**2            # Y = x^2
y2 = np.sin(x)       # Y = sin(x)
y3 = np.exp(x)       # Y = e^x
y4 = np.log(np.abs(x) + 1)  # Y = log(|x| + 1)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot each function
plt.plot(x, y1, label='Y = x^2')
plt.plot(x, y2, label='Y = sin(x)')
plt.plot(x, y3, label='Y = e^x')
plt.plot(x, y4, label='Y = log(|x| + 1)')

# Adding title and labels
plt.title("Multiple Function Plots")
plt.xlabel("x values")
plt.ylabel("y values")

# Show grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
