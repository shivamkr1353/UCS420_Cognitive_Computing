```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
M = float(input("Enter a value for M: "))
x = np.linspace(-10, 10, 400)
y1 = M * x**2        # y = M * x^2
y2 = M * np.sin(x)   # y = M * sin(x)
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label=f'y = {M} * x^2', color='b', linestyle='-', linewidth=2)
plt.plot(x, y2, label=f'y = {M} * sin(x)', color='r', linestyle='--', linewidth=2)
plt.legend()
plt.grid(True)
plt.title(f'Plot of y = {M} * x^2 and y = {M} * sin(x)', fontsize=14)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Art'],
    'Score': [85, 90, 78, 88, 92]
}
df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
bar_plot = sns.barplot(x='Subject', y='Score', data=df, palette='Set2')
for p in bar_plot.patches:
    bar_plot.annotate(format(p.get_height(), '.1f'),
                      (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha='center', va='center', fontsize=12, color='black',
                      xytext=(0, 5), textcoords='offset points')
plt.title('Scores by Subject', fontsize=16)
plt.xlabel('Subject', fontsize=14)
plt.ylabel('Score', fontsize=14)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
roll_number = 102497026
np.random.seed(roll_number)
data = np.random.randn(50)
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].plot(np.cumsum(data), label='Cumulative Sum', color='b')
axes[0].set_title('Cumulative Sum')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Cumulative Value')
axes[0].grid(True)
axes[1].scatter(np.arange(len(data)), data, label='Random Noise', color='r')
axes[1].set_title('Scatter Plot with Random Noise')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Random Value')
axes[1].grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
url = 'https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv'
data = pd.read_csv(url)
data.replace([np.inf, -np.inf], np.nan, inplace=True)
data.dropna(inplace=True)
print(data.head())
plt.figure(figsize=(10, 6))
sns.lineplot(x=data['month_number'], y=data['total_profit'])
plt.title('Total Profit Over Months')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
plt.figure(figsize=(10, 6))
sns.lineplot(data=data[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']])
plt.title('Product Sales Over Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Products')
plt.show()
plt.figure(figsize=(12, 8))
data.drop(columns='month_number').mean().plot(kind='bar', color='skyblue')
plt.title('Bar Chart for All Features (Mean Values)')
plt.ylabel('Mean Value')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
